```python
names=[]
prices=[]
rating=[]
sim=[]
processor=[]
ram=[]
battery=[]
display=[]
camera=[]
card=[]
os=[]

for i in containers:
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span', {'class':'price'}).text)
    except:
        price.append(np.nan)
    try:
        score_classes = ["score rank-1-bg", "score rank-2-bg", "score rank-3-bg", "score rank-4-bg", "score rank-5-bg"]
        rating_found = False
        for score_class in score_classes:
            score_div = i.find('div', {'class': score_class})
            if score_div:
                rating.append(score_div.find('b').text)
                rating_found = True
                break
        if not rating_found:
            rating.append(np.nan)
    except:
        rating.append(np.nan)
    x= i.find('ul', {'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
        




```

```python
len(os)
```

```python
df= pd.DataFrame({
    "model":names,
    "price":prices,
    "rating":rating,
    "sim":sim,
    "processor":processor,
    "ram":ram,
    "battery":"battery",
    "display":display,
    "camera":camera,
    "card":card,
    "os":os
})
```
